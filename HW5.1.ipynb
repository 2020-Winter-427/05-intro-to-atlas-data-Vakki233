# Intro to ATLAS Data

```python
import numpy as np
import uproot
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Get access of the sample file and the key names
file1 = uproot.open("mc_361106.Zee.1largeRjet1lep.root")
file1.keys()
```

```python
tree1 = file1["mini"].pandas.df(flatten = False)
tree1
```

```python
#Plot the number of Lepton histogram
data1 = tree1["lep_n"]
plt.figure(figsize=(10,8))
plt.hist(data1,bins=3)
plt.title("Number of Lepton",fontsize=20)
plt.xlabel("#Lepton")
plt.ylabel("Counts")
plt.show()
```

```python
#Plot the normalized number of lepton histogram
plt.figure(figsize=(10,8))
plt.hist(data1,bins=3,density=True)
plt.title("Normalized Number of Lepton",fontsize=20)
plt.xlabel("#Lepton")
plt.ylabel("Counts")
plt.show()
```
